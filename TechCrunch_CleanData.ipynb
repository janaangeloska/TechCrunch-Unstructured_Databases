Рачно задавање на имиња на колони и читање без заглавие

```python
import pandas as pd

column_auth = ['auth_id', 'name', 'meibi', 'meibix', 'avg_words', 'avg_words_len_nostop']
authors_df = pd.read_csv('archive/authors.csv', header=None, names=column_auth, on_bad_lines='skip')

column_com = ['com_id', 'post_id', 'content', 'author', 'date', 'vote']
comments_df = pd.read_csv('archive/comments.csv' , header=None, names=column_com, on_bad_lines='skip')

column_inl = ['inl_id', 'post_id', 'title', 'author', 'date', 'URL']
inlinks_df = pd.read_csv('archive/inlinks.csv',  header=None, names=column_inl, on_bad_lines='skip')

column_post = ['post_id', 'title', 'bloggers_name', 'bloggers_id', 'nr_of_comments', 'content', 'URL', 'date', 'nr_retrived_inlinks', 'nr_retrived_comments', 'post_len', 'post_len_nostop', 'avg_post_len', 'avg_post_len_nostop', 'MEIBI_score', 'MEIBIX_score']
posts_df = pd.read_csv('archive/posts.csv', header=None, names=column_post, on_bad_lines='skip')
```

Справување со нулеви вредности

```python
print("Authors null values:")
print(authors_df.isnull().sum())
```

```python
print("Posts null values:")
print(posts_df.isnull().sum())
```

```python
print("Comments null values:")
print(comments_df.isnull().sum())
```

```python
print("Inlinks null values:")
print(inlinks_df.isnull().sum())
```

 Пополнување на атрибути без кои може да постои објавата

```python
posts_df['title'] = posts_df['title'].fillna('unknown')
posts_df['content'] = posts_df['content'].fillna('unknown')
```

```python
print("Posts null values:")
print(posts_df.isnull().sum())
```

Бришење на коментарите за кои не е познат авторот

```python
comments_df = comments_df.dropna(subset=['content', 'author'])
```

```python
print("Comments null values:")
print(comments_df.isnull().sum())
```

```python
inlinks_df['author'] = inlinks_df['author'].fillna('unknown')
inlinks_df['title'] = inlinks_df['title'].fillna('unknown')
```

```python
print("Inlinks null values:")
print(inlinks_df.isnull().sum())
```

Проверка на дупликат автори

```python
print(authors_df['auth_id'].duplicated().sum())
```

Валидација на врските помеѓу датотеките

```python
valid_authors = posts_df['bloggers_id'].isin(authors_df['auth_id'])
print(f"Валидни автори во posts_df: {valid_authors.sum()} / {len(posts_df)}")
```

```python
valid_posts_comments = comments_df['post_id'].isin(posts_df['post_id'])
print(f"Валидни постови во comments_df: {valid_posts_comments.sum()} / {len(comments_df)}")
```

```python
valid_posts_inlinks = inlinks_df['post_id'].isin(posts_df['post_id'])
print(f"Валидни постови во inlinks_df: {valid_posts_inlinks.sum()} / {len(inlinks_df)}")
```
